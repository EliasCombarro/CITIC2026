## Resolviendo un problema sencillo (Ising)

```python
import dimod
J = {(0,1):1, (0,2):1}
h = {}
problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)
print("Nuestro problema es:")
print(problem)
```

```python
from dwave.system import DWaveSampler
from dwave.system import EmbeddingComposite
sampler = EmbeddingComposite(DWaveSampler())
result = sampler.sample(problem, num_reads=10)
print("Las soluciones obtenidas son:")
print(result)
```

## Resolviendo un problema sencillo (QUBO)

```python
J = {(0,1):1, (0,2):1}
h = {0:-2, 2:-1}
problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.BINARY)
print("Nuestro problema es:")
print(problem)
```

```python
result = sampler.sample(problem, num_reads=10)
print("Las soluciones obtenidas son:")
print(result)
```

## Forzando el sistema

```python
def grafo_circular(N):
  J = {}
  for i in range(N-1):
    J[(i,i+1)] = 1
  J[(N-1,0)] = 1
  h = {}
  problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)
  return problem
```

```python
result = sampler.sample(grafo_circular(10), num_reads=10)
print("Las soluciones obtenidas son:")
print(result)
```

```python
result = sampler.sample(grafo_circular(11), num_reads=10)
print("Las soluciones obtenidas son:")
print(result)
```
